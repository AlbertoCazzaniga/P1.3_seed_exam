```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
res = np.zeros([38,10,3])
x = np.append(np.arange(1000, 35001, 1000), np.array([50000,70000,100000]))

l = list(range(1000, 35001, 1000))
l.append(50000)
l.append(70000)
l.append(100000)
```

```python
basename = "./results/results_"
for i, it in enumerate(l):
    res[i,:] = np.loadtxt(basename + str(it) + ".dat")
    
```

```python
res_mean = np.mean(res, axis = 1)
res_std = np.std(res, axis = 1)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
plt.title('Runtime find() function for maximally unbalanced tree (1000 keys)', fontsize=15)
ax.yaxis.grid(which="major", color='black', linestyle='-', linewidth=1, alpha = 0.3)
#ax.xaxis.grid(which="major", color='black', linestyle='-', linewidth=1, alpha = 0.3)
plt.errorbar(x, res_mean[:,0], res_std[:,0], color='blue', marker='o', label = 'exec. time')
plt.fill_between(x,res_mean[:,0] + res_std[:,0], res_mean[:,0] - res_std[:,0], color = 'blue', alpha=0.3, label = 'std_dev')
plt.legend(frameon=True, loc=2, fontsize=12)
plt.xlabel("tree size", fontsize=12)
plt.ylabel("runtime (ms)", fontsize=12)
plt.savefig("./plots/unbalanced.png")
```

```python
fig,ax = plt.subplots(figsize=(10,8))
plt.title('Runtime find() function for randomly balanced tree (1000 keys)', fontsize=15)
ax.yaxis.grid(which="major", color='black', linestyle='-', linewidth=1, alpha = 0.3)
plt.fill_between(x,res_mean[:,1] + res_std[:,1], res_mean[:,1] - res_std[:,1], color = 'green', alpha=0.3)
plt.errorbar(x, res_mean[:,1], res_std[:,1], color='green', marker='o', label = 'exec. time tree')
plt.fill_between(x,res_mean[:,2] + res_std[:,2], res_mean[:,2] - res_std[:,2], color = 'orange', alpha=0.3)
plt.errorbar(x, res_mean[:,2], res_std[:,2], color='orange', marker='o', label = 'exec. time std::map')
plt.legend(frameon=True, loc=2, fontsize=15)
plt.xlabel("tree size", fontsize=12)
plt.ylabel("runtime (ms)", fontsize=12)
plt.savefig("./plots/rand_bal.png")
```
